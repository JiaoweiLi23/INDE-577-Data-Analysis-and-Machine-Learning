# Principal Component Analysis (PCA) 
Principal Component Analysis is a statistical technique used in machine learning and data science for dimensionality reduction while preserving as much variability in the data as possible. It works by identifying the directions, called principal components, along which the variation in the data is maximum. PCA transforms the original data to a new coordinate system where the greatest variances lie on the first coordinates (the principal components), and each subsequent component has the highest variance possible under the constraint that it is orthogonal to the preceding components. This technique is particularly useful in processing and visualizing high-dimensional data, aiding in uncovering hidden patterns and reducing computational costs. However, it's important to note that PCA is a linear method and may not capture complex relationships in data as effectively as some non-linear dimensionality reduction techniques.
Certainly! Principal Component Analysis (PCA) is a fundamental technique in data science and machine learning for dimensionality reduction and data visualization. Here's a more detailed explanation:

### 1. Concept

1. Purpose
   - PCA reduces the dimensionality of data, simplifying the dataset while retaining as much information as possible. It's commonly used when you have many correlated variables and need to focus on the most important ones.

2. Principal Components
   - Principal Components (PCs) are new, artificial features created by PCA. They are linear combinations of the original variables and are orthogonal (perpendicular) to each other. The first principal component captures the most variance in the data, the second captures the second most, and so on.

3. Variance and Information 
   - The core idea is that PCs with higher variance capture more 'information.' PCA transforms the data to a new coordinate system where the greatest variances (information) are on the first axes and decrease with each subsequent axis.

### 2. Steps 

1. Standardize the Data
   - PCA is affected by the scale of the features, so it's common to standardize the data first (mean of 0 and a standard deviation of 1).
  $$z = \frac{(x - \mu)}{\sigma} $$

Where:
- $ x $ is the original value of the data point.
- $ \mu$ is the mean of the feature.
- $\sigma $ is the standard deviation of the feature.
- $ z $ is the standardized value.
Finally, we form the $m\times n$ matrix $A$. 

2. Covariance Matrix Computation
   - Calculate the covariance matrix to understand how variables in the data are related. In the context of PCA, a covariance matrix helps to identify which dimensions are correlated.
   $$
S = \frac{1}{n-1}AA^T
$$

3. Compute Eigenvalues and Eigenvectors
   - Eigenvalues and eigenvectors are computed from the covariance matrix. Eigenvectors give the directions of the new feature space, and eigenvalues give the magnitude (or variance captured by each principal component).
   - These eigenvectors are columns of the matrix $U$ in the singular value decomposition of $A$, up to the factor $n-1$. Further, we denote the eigenvalues by $\sigma_{i}^{2}$. This is equivalent to the **Singular Value Decomposition** of our shifted training set matrix $A$,

$$
A = \mathbf{U} \mathbf{\Sigma} \mathbf{V}^{T}, 
$$


4. Sort and Select Principal Components
   - Eigenvectors are sorted by eigenvalues in descending order. This sorting gives the components in order of significance. Here, you decide how many principal components to keep. Often, this is based on a threshold of the total variance explained.

5. Transform the Original Dataset
   - Finally, the original data is transformed into this new subspace using the selected principal components. This transformation results in new, uncorrelated features (the principal components).

### 3. Considerations
  
- **Limitations**: PCA assumes linearity and relies on orthogonal transformations. It may not perform well with nonlinear data relationships. Also, PCA can make the data less interpretable since the principal components are combinations of the original features.

- **Choosing the Number of Components**: The number of components to keep is a decision point in PCA. Often, a scree plot is used, where we look for the 'elbow' in the graph of explained variance similar like in k-means.

### 4. Implementation

#### 4.1 Dataset
In this section, we utilized a publicly accessible dataset from Kaggle, originating from a continuous cardiovascular research project conducted in Framingham, Massachusetts. This dataset's primary objective is to forecast the 10-year risk of coronary heart disease (CHD) in patients. It contains detailed information on patients, encompassing more than 4,000 entries and 15 different attributes.

```python
from collections import Counter
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
data = pd.read_csv("framingham.csv")
data=data.dropna()
data.head(10)
```

To begin, we must standardize the dataset using the following code:

```python
X = data.iloc[:,1:15]
X_scaler = StandardScaler()
X = pd.DataFrame(X_scaler.fit_transform(X))
y = data["TenYearCHD"]
```

Secondly, following the steps in section two, we will compute the Covariance Matrix.

```python
cov_matrix = np.cov(X, rowvar=False)
```

Step 3: Calculate Eigenvalues and Eigenvectors

```python
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
```

Step 4: Sort Eigenvalues and Eigenvectors

```python
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]
```

Next, step 5, select the top k eigenvectors, here we choose to select two principal components.

```python
k = 2 # Number of principal components
pca_components = sorted_eigenvectors[:, :k]
```

Step 6: Transform the original dataset
X_pca is the data transformed into the new principal component space

```python
X_pca = np.dot(X, pca_components)
```

Excellent work! Now, let's visualize the results of our PCA algorithm.

```python
def colors(y):
    if y == 1:
        return "red"
    elif y == 0:
        return "blue"
    else:
        return "green"

c = [colors(label) for label in y]

plt.figure(figsize = (10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c = c)
plt.xlabel("First Principle Component", fontsize = 15)
plt.ylabel("Second Principle Component")
plt.show()
```

Instead of building a PCA algorithm from scratch, we can also conveniently utilize the built-in function in the scikit-learn package in Python for feature selection. The following code demonstrates how to use this package to perform PCA.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# Since we have already standardized our dataset at the beginning, we will proceed with PCA from step 2.
# Step 2: Create a PCA instance
# We can specify the number of components we want, for example, 2
# Here, as we aim to visualize all the components, we will not specify the number of components when applying PCA.
pca = PCA()

# Step 3: Fit PCA on the standardized data
pca.fit(X)

# Step 4: Transform the data according to the fitted PCA
X_pca = pca.transform(X)

# X_pca now contains the principal components

# If you want to create a DataFrame of the principal components:
pca_df = pd.DataFrame(data=X_pca)

# Optional: Check the explained variance ratio
explained_variance = pca.explained_variance_ratio_
print(explained_variance)

# Explained variance tells us how much information (variance) is captured by each principal component
```

```python
# Calculate the variance ratios
variance_ratios = pca.explained_variance_ratio_

# Create the scree plot
plt.figure(figsize=(8, 4))
plt.bar(range(1, len(variance_ratios) + 1), variance_ratios, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(1, len(variance_ratios) + 1), np.cumsum(variance_ratios), where='mid', label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

### 5. Conclusion
Principal Component Analysis (PCA) is a powerful statistical technique in data analysis and machine learning for dimensionality reduction. It is particularly useful in processing and visualizing high-dimensional data. By transforming the data into a set of orthogonal (uncorrelated) components that capture the most variance, PCA simplifies the complexity of high-dimensional data while retaining essential information.

Key strengths of PCA include its ability to uncover hidden patterns in the data, reduce noise, and improve the efficiency of other machine learning algorithms by reducing the number of features. It's often used as a preliminary step in data preprocessing before applying complex algorithms, aiding in visualization, and sometimes even improving model performance due to the removal of redundant features.

However, PCA has limitations. It assumes linearity and is less effective in handling non-linear relationships. The interpretation of principal components can sometimes be less intuitive than the original features, and the technique is sensitive to the scaling of data, necessitating careful preprocessing.

In conclusion, PCA is an invaluable tool in the data scientist's arsenal, offering a balance between data simplification and information retention. Its application can lead to insightful discoveries and more efficient processing in various data-driven fields, though it should be applied with an understanding of its assumptions and limitations.
