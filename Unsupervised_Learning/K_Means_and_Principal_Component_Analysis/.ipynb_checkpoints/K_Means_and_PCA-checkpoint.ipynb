# K-Means Clustering

K-means clustering is a popular unsupervised machine learning algorithm used for partitioning a dataset into a set of distinct, non-overlapping subgroups, or 'clusters'. It works by assigning data points to clusters based on their proximity to the nearest cluster center, with each cluster represented by a centroid. The algorithm iteratively updates the position of these centroids to minimize the within-cluster variance, essentially finding the most central point in each cluster. It's widely used for data analysis in various fields due to its simplicity and efficiency, particularly suitable for large datasets. However, it requires specifying the number of clusters beforehand and can be sensitive to the initial placement of centroids.
Certainly! K-means is a straightforward yet powerful algorithm for clustering data into distinct groups. Here's a step-by-step explanation:

### 1. Algorithm
#### 1.1 Goal
   - **Objective**: The primary goal of K-means is to group similar data points together and discover underlying patterns. To achieve this, K-means looks for a fixed number (k) of clusters in a dataset.
   - **Clusters**: A cluster refers to a collection of data points aggregated together because of certain similarities.

#### 1.2 Choose the Number of Clusters (K)
   - We need to decide the number of clusters (k) in advance.
   - This can be based on prior knowledge about the data, or you can use methods like the Elbow Method to determine an optimal value for k.

#### 1.3 Initialize Centroids
   - Randomly select k data points as the initial centroids. Centroids are the center points of each cluster.
   - The choice of initial centroids can affect the final clustering result.

#### 1.4. Assign Data Points to the Nearest Centroid
   - For each data point, find the nearest centroid (using a distance measure like Euclidean distance) and assign the point to that cluster.
       - In this Jupyter notebook, we will use Euclidean distance to find the neares centroid
  $$ D = \sqrt{\sum_{i=1}^{n} (q_i - p_i)^2} $$
   - A data point is considered to be in a particular cluster if it is closer to that cluster's centroid than any other centroid.

#### 1.5 Recompute Centroids
   - Once all points are assigned to clusters, recalculate the centroids for each cluster. This new centroid is the mean of all points in the cluster.
   - Essentially, the centroid of a cluster is the center of all the data points in that cluster.

#### 1.6 Iterative Optimization
   - Repeat the steps of assigning data points and recalculating centroids until certain criteria are met (e.g., centroids do not change significantly, or a maximum number of iterations is reached).
   - This iterative process optimizes the positions of the centroids to create clusters with minimal intra-cluster variance.

#### 1.7 Final Clusters
   - The algorithm ends when the centroids have stabilized, and the assignment of data points to clusters no longer changes.
   - These final clusters are the output of the K-means algorithm.

### 2. Considerations
   - **Choosing k**: Deciding the number of clusters can be non-trivial and may require domain knowledge or methods like the Elbow Method.
   - **Sensitivity to Initialization**: The initial position of centroids can impact the final outcome. Sometimes K-means is run multiple times with different initializations.
   - **Handling Different Scales**: Standardization of data may be necessary, as K-means can be sensitive to the scale of the data.
   - **Not Suitable for All Data Types**: K-means works well for spherical clusters but may not perform well for clusters of different shapes and sizes.

### 3. Elbow Method 
The Elbow Method is a technique used in cluster analysis, specifically in **determining the number of clusters (k) in K-means clustering**. This method involves running the K-means clustering algorithm on the dataset for a range of values of k (for example, k from 1 to 10), and then for each value of k, computing the sum of squared distances from each point to its assigned center. When these overall dispersions are plotted against the number of clusters, the resulting curve often has an "elbow point," which is considered to be an indicator of the appropriate number of clusters.

#### 3.1 Steps

1. **Choose a Range of k**: Decide on a range of k values to evaluate. 

2. **Run K-means for Each k**: For each k value in the range, run the K-means clustering algorithm on our dataset.

3. **Calculate Sum of Squared Distances**: For each k, calculate the total within-cluster sum of square (WSS). This is the sum of the squared distance between each data point and its closest centroid.

$$ \text{WSS} = \sum_{i=1}^{k} \sum_{x \in C_i} ||x - \mu_i||^2 $$

Where:
- $ k $ is the number of clusters.
- $ C_i $ represents the set of all points in the $ i^{th} $ cluster.
- $x $ is a data point in cluster $ C_i $.
- $\mu_i $ is the centroid of the $ i^{th} $ cluster.
- $ ||x - \mu_i|| $ is the Euclidean distance between a data point $ x $ and the centroid $ \mu_i $ of its cluster.


4. **Plot the Results**: Create a plot of the number of clusters (k) against the corresponding WSS. 

5. **Identify the Elbow Point**: Look at the plot and locate the "elbow point," where the rate of decrease sharply changes. This point indicates a suitable number of clusters for your data. The idea is that adding another cluster beyond this number doesnâ€™t give much better modeling of the data.

### 3. Implementation
#### 3.1 Dataset
In this analysis, we used a dataset 'palmer penguins' that is publicly available on Kaggle, originally collected and provided by Dr. Kristen Gorman and the Palmer Station, Antarctica LTER, part of the Long Term Ecological Research Network. For our study, we selected 'species' as the dependent variable (y) and two features: 'culmen_length_mm' and 'culmen_depth_mm'. The species comprises three categories: Adelie, Chinstrap, and Gentoo.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

data= pd.read_csv("palmer_penguins.csv")
data=data.dropna()
y = data.species.values
encoder = LabelEncoder()
y = encoder.fit_transform(y)
X = data[['bill_length_mm','bill_depth_mm']].to_numpy()
x=X
```

```python
plt.figure(figsize=(10, 8))
for label in set(y):
    indices = [i for i, l in enumerate(y) if l == label]
    if label == 1:
        plt.scatter(x[indices, 0], x[indices, 1], c="red", label="Adelie")
    elif label == 2:
        plt.scatter(x[indices, 0], x[indices, 1], c="blue", label="Chinstrap")
    else:
        plt.scatter(x[indices, 0], x[indices, 1], c="yellow", label="Gentoo")
plt.xlabel("bill_length_mm", fontsize=15)
plt.ylabel("bill_depth_mm", fontsize=15)
plt.grid(True)
plt.legend()
plt.show()
```

First, let's arbitrarily choose k=3 and randomly select 3 data points as the initial centroids. We can run the following code to choose the initial centroids:

```python
centers = []
for k in range(3):
    i = np.random.randint(0, 100)
    point = (X[i, 0], X[i, 1], k)
    centers.append(point)
plt.figure(figsize = (10, 8))
plt.scatter(X[:, 0], X[:, 1])
plt.scatter([centers[0][0]], [centers[0][1]], color = "red", label = centers[0][2],s=100)
plt.scatter([centers[1][0]], [centers[1][1]], color = "lightseagreen", label = centers[1][2],s=100)
plt.scatter([centers[2][0]], [centers[2][1]], color = "magenta", label = centers[2][2],s=100)
plt.legend()
plt.show()
```

Next, we need to define the functions required for implementing K-means.

```python
# calculate distance between each point
def distance(i, X, center):
    return np.sqrt(sum((X[i, j] - center[j])**2 for j in range(2)))
# assign data points to the nearest centroid
def assign_label(i, X, centers):
    distances = np.array([distance(i, X, center) for center in centers])
    label = np.argmin(distances)
    return label 
def assign_clusters(X, centers):
    return [assign_label(i, X, centers) for i in range(len(X))]
```

```python
def colors(y):
    if y == 0:
        return "red"
    elif y == 1:
        return "blue"
    elif y == 2:
        return "yellow"
    else:
        return "black"
c = [colors(label) for label in y]
current_labels = assign_clusters(X, centers)
current_colors = [colors(label) for label in current_labels]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 10))
ax1.scatter(X[:, 0], X[:, 1], c = c)

ax2.scatter(X[:, 0], X[:, 1], c = current_colors)
ax2.scatter([centers[0][0]], [centers[0][1]], color = "red", label = centers[0][2],s=100)
ax2.scatter([centers[1][0]], [centers[1][1]], color = "lightseagreen", label = centers[1][2],s=100)
ax2.scatter([centers[2][0]], [centers[2][1]], color = "magenta", label = centers[2][2],s=100)

plt.show()
```

The plot displayed above illustrates the results of the clustering. Now, after all points are assigned to clusters, we need to recalculate the centroids for each cluster. This new centroid is the mean of all points in the cluster.

```python
def update_centers(X, centers):
    current_labels = assign_clusters(X, centers)
    k = len(centers)
    center_means = {i : [0, 0] for i in range(k)}
    for i in range(len(X)):
        center_means[current_labels[i]][0] += X[i, 0]
        center_means[current_labels[i]][1] += X[i, 1]

    for i in range(k):
        num_occur = current_labels.count(i)
        center_means[i][0] = center_means[i][0]/num_occur
        center_means[i][1] = center_means[i][1]/num_occur
    
    return [(center_means[i][0], center_means[i][1]) for i in center_means]
        
```

Now, we can define our K-means clustering function using the functions defined above:

```python
def k_means_clustering(X, k, max_iter = 1000):
    centers = []
    for j in range(k):
        i = np.random.randint(0, 100)
        point = (X[i, 0], X[i, 1], j)
        centers.append(point)
    
    for _ in range(max_iter):
        centers = update_centers(X, centers)
    
  
    return centers
```

```python
def calculate_wss(X):  
    d = 0
    for i in range(len(X)):
        d += distance(i, X, centers[current_labels[i]])
    return d
```

```python
centers = k_means_clustering(X, 5)
current_labels = assign_clusters(X, centers)
current_colors = [colors(label) for label in current_labels]
dis=calculate_wss(X)
```

Since we have built our k_means algorithm, then we need to find appropriate number of k by using the Elbow Method. First, we need to define our elbow method function.

```python
wss = []
k_range = [2,3,4,5,6,7]
for kk in k_range:
    centers = k_means_clustering(X, kk)
    current_labels = assign_clusters(X, centers)
    current_colors = [colors(label) for label in current_labels]
    dis=calculate_wss(X)
    wss.append(dis)
plt.figure(figsize=(10,6))
plt.plot(k_range, wss, 'bo-')
plt.title("Elbow Method For Optimal k")
plt.xlabel("Number of clusters")
plt.ylabel("Within-Cluster Sum of Square")
plt.show()
```

The elbow method plot indicates that the most appropriate number of clusters (k) should be 4, as the rate of decrease sharply changes at this point. However, given our prior knowledge of the dataset, which includes three species categories, we will still choose k=3.

```python
centers = k_means_clustering(X, 3,max_iter=10000)
current_labels = assign_clusters(X, centers)
current_colors = [colors(label) for label in current_labels]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 10))
ax1.scatter(X[:, 0], X[:, 1], c = c)
ax1.set_title("Original Data")
ax2.scatter(X[:, 0], X[:, 1], c = current_colors)
ax2.set_title("K-means Data")
plt.show()
```

### 4. Conclusion
K-means is a fundamental clustering technique in data science, useful for quick and intuitive grouping of data, but its effectiveness can depend on the nature of the dataset and the correct choice of parameters. The algorithm is efficient and works well with large datasets, but it has some limitations:

- The number of clusters, k, must be specified in advance.
- It's sensitive to the initial choice of centroids, which can lead to different clustering results.
- K-means assumes that clusters are spherical and evenly sized, which might not always be the case in real-world data.
- It is not well-suited for clusters of different densities or non-convex shapes.

The Elbow Method is often used to determine an appropriate value for k, where the "elbow" point in a plot of the within-cluster sum of squares against the number of clusters indicates an optimal cluster number.
