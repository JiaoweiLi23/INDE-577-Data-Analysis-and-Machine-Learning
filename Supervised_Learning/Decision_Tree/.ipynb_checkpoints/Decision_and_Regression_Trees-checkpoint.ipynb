# Decision and Regression Trees
A decision tree is a flowchart-like tree structure used in machine learning and data mining. It's a popular tool for classification and regression tasks. The tree is composed of nodes that represent decisions or splits, and branches that represent outcomes of these decisions.

# Decision Tree
## 1. Components of a Decision Tree
* Root Node: The topmost node, representing the entire dataset. It splits into two or more homogeneous sets.
* Splitting: Dividing a node into two or more sub-nodes based on a certain condition.
* Decision Node: When a sub-node splits into further sub-nodes.
* Leaf/Terminal Node: Nodes with no further division, representing the output.
* Pruning: Removing sub-nodes from a decision node to prevent overfitting.
* Branch/Sub-Tree: A subsection of the entire tree.

## 2. Procedures
* Select the Best Attribute: Use a criterion like Information Gain, Gini Index, etc., to find the best attribute to split the data.
* Split the Data: Based on the selected attribute, split the dataset into subsets.
* Repeat for Sub-Nodes: Repeat the process for each sub-node, choosing the best attribute to split further until a stopping criterion is reached (e.g., all data in a node belong to the same class, no more attributes, etc.).
* Classify the Outcome: Each leaf node represents a classification or decision.

## 3. Advantages and Disadvantages
* Advantages: Easy to understand and interpret, requires little data preparation, and can handle both numerical and categorical data.
* Disadvantages: Prone to overfitting, can be unstable with small changes in data, and may create biased trees if some classes dominate.

## 5. Implementation
### Dataset Visualization
In this analysis, we used a dataset 'palmer penguins' that is publicly available on Kaggle, originally collected and provided by Dr. Kristen Gorman and the Palmer Station, Antarctica LTER, part of the Long Term Ecological Research Network. For our study, we selected 'Sex' as the dependent variable (y) and two features: 'culmen_length_mm' and 'culmen_depth_mm'.

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.tree import export_text
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 

data= pd.read_csv("palmer_penguins.csv")
data=data.dropna()
data=data[(data['species']=='Gentoo')]
y = data.sex.values
y = np.where(y == 'Female', 0, 1)
X = data[['bill_length_mm','bill_depth_mm']]
````

````python
x=X.values
colors = ["red" if label == 1 else "blue" for label in y]
plt.figure(figsize=(10, 8))
plt.scatter(x[:, 0], x[:, 1], c = colors)
plt.xlabel("feature bill_length_mm", fontsize = 15)
plt.ylabel("feature bill_depth_mm", fontsize = 15)
plt.grid(True)
plt.show()
````

The plot above displays penguins of the Gentoo species. The blue points represent males, and the red points represent females, with respect to two features: bill length and bill depth.

Next, we utilize the built-in 'train_test_split' function in scikit-learn to divide the data into training and testing sets.

````python
# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
x=X_train
colors = ["red" if label == 1 else "blue" for label in y_train]
plt.figure(figsize=(10, 8))
plt.scatter(x[:, 0], x[:, 1], c = colors)
plt.xlabel("feature bill_length_mm", fontsize = 15)
plt.ylabel("feature bill_depth_mm", fontsize = 15)
plt.grid(True)
plt.show()
````

We then applied a Decision Tree classifier to our training set, followed by using the test set to evaluate the accuracy of our predictions. The result shows that our classifier accuracy is 88.9%. This can be shown by running the following code cell.

````python
# Create a Decision Tree Classifier
clf = DecisionTreeClassifier()

# Train the classifier
clf.fit(X_train, y_train)

# Predict the response for test dataset
y_pred = clf.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
````

### Results Visualization
Now that our model has been trained we can visualize the tree structure of our current model by calling the ```sklearn.tree.plot_tree()``` function.

````python
tree_rules = export_text(clf,
                         feature_names = ["x_0", "x_1"])
print(tree_rules, "\n")


plt.figure(figsize=(20, 8))
a = plot_tree(clf,
                   feature_names = ["x_0", "x_1"],
                   class_names = ["red", "blue"],
                   rounded = True,
                   filled = True, 
                   fontsize=14)
plt.show()
````

Then We can visualize the decision regions generated by our trained decision tree by running the following code cell. 

````python
from mlxtend.plotting import plot_decision_regions
X_test=np.array(X_test)
plt.figure(figsize = (10, 8))
plot_decision_regions(X_test, y_test, clf = clf)
plt.xlabel("feature x_0", fontsize = 15)
plt.ylabel("feature x_1", fontsize = 15)
plt.grid()
plt.show()
````

### Evaluation of the Model
Next we evaluate our models performance on the testing data. 

A confusion matrix is a tool often used in machine learning to visualize the performance of a classification algorithm. It's a table that allows you to assess the accuracy of a classifier by comparing the actual values (true outcomes) with the values predicted by the model.

This is done by running the following code cells. 

````python
# predicted values on the testing data
test_pred_decision_tree = clf.predict(X_test)

# Import metrics from sklearn 
from sklearn import metrics

# Note: visualizing your tree above will be weird after running seaborn
import seaborn as sns

# The confusion matrix
confusion_matrix = metrics.confusion_matrix(y_test, test_pred_decision_tree)

# Convert confusion matrix into dataframe
matrix_df = pd.DataFrame(confusion_matrix)


plt.figure(figsize=(10, 8))
ax = plt.axes()
sns.set(font_scale=1.3)

sns.heatmap(matrix_df,
            annot = True, 
            fmt = "g",
            ax = ax, 
            cmap = "magma", 
            cbar = False)

ax.set_title("Confusion Matrix - Decision Tree")
ax.set_xlabel("Predicted Label", fontsize=15)
ax.set_xticklabels(["red", "blue"])
ax.set_ylabel("True Label", fontsize=15)
ax.set_yticklabels(["red", "blue"], rotation=0)
plt.show()
````

````python
from sklearn.metrics import classification_report

y_pred = clf.predict(X_test)

#Checking performance our model with classification report.
print(classification_report(y_test, y_pred))
````

# Regression Tree
A regression tree is a type of decision tree used for predicting continuous numerical values, rather than categorical labels as in classification trees. They are particularly useful in scenarios where the relationship between the features and the target variable is non-linear. 

In a regression tree, each leaf of the tree represents a numerical value. When a new data point is fed into the tree, it traverses the tree based on the decision rules until it reaches a leaf. The value of this leaf is the prediction for the input data point.

The tree is constructed by splitting the training set into subsets. These splits are made at nodes based on a certain condition on a specific feature.

Unlike classification trees, which use criteria like Gini impurity or entropy, regression trees use different metrics like the mean squared error (MSE) to decide the best split.

## 1. Regression Tree Construction
* Starting from the root, the dataset is split recursively.
* At each step, the feature and the threshold that give the best split (in terms of reducing MSE) are chosen.
* This process continues until a stopping criterion is reached, which could be a maximum depth of the tree, a minimum number of samples in a leaf, or a minimal improvement in the error.

## 2. Prediction
* For a new data point, the tree is traversed by testing the feature values against the conditions in the nodes until a leaf node is reached.
* The predicted value is the mean target value of the training samples in that leaf.

## 3. Advantages and Disadvantages
**Advantages:**
1. Easy to understand and interpret.
2. Can model complex non-linear relationships.
3. Requires little data preprocessing.

**Disadvantages:**
1. Prone to overfitting, especially with complex trees.
2. Sensitive to changes in the training data.
3. Can struggle with extrapolation outside the range of the training data.

## 4. Implementation
### Dataset Visualization
In this analysis, we used a dataset 'California Housing' that is publicly available in python sklearn.datasets. 

**Features**: The dataset includes several features (variables) for each block group:

*Median Income*: The median income of households within a block.

*Housing Median Age*: The median age of the houses.

*Average Rooms*: The average number of rooms in houses.

*Average Bedrooms*: The average number of bedrooms in houses.

*Population*: The total number of people residing within a block.

*Average Occupancy*: The average number of occupants per household.

*Latitude*: The geographical latitude of the block.

*Longitude*: The geographical longitude of the block.

**Target Variable**: The main target variable is the median house value for California districts. This is what models typically aim to predict based on the other features.

````python
from sklearn.datasets import fetch_california_housing
# Target is median house value 
X, y = fetch_california_housing(return_X_y = True)
````

### Tree Visualization
We use function DecisionTreeRegressor in sklearn.tree to perform Regression Tree.

````python
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
reg = DecisionTreeRegressor(max_depth=3)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 12)
reg.fit(X_train, y_train)
plt.figure(figsize=(20,8))
a = tree.plot_tree(reg,
                   feature_names = None,
                   class_names = None,
                   rounded = True,
                   filled = True,
                   fontsize = 14)
plt.show()
````

### Evaluation
The 'max_depth' Key represents the maximum depth of the tree. Deeper trees can model more complex patterns but can also lead to overfitting.
In this section, we performed regression tree analysis with varying maximum depths and evaluated their Mean Squared Error (MSE).

````python
MSE=[]
for i in range(1,20):
    reg1 = DecisionTreeRegressor(max_depth=i, random_state = 12)
    reg1.fit(X_train, y_train)
    y_pred = reg1.predict(X_test)
    MSE.append(mean_squared_error(y_test, y_pred))
````

````python
plt.figure(figsize=(8, 4))
plt.plot(np.arange(1, 20), MSE)
plt.xlabel("max_depth", fontsize = 15)
plt.ylabel("MSE", fontsize = 15)
plt.title("MSE for different max_depth", fontsize = 18)
plt.show()
````

From the plot above, we observe that the MSE (Mean Squared Error) increases when the max_depth is either too small or too large. The minimum MSE is achieved when the max_depth is close to 10, at which point the MSE is approximately 0.4.
